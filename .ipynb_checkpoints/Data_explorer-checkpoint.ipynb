```python
import os
from pathlib import Path
import data_loader_utils
import itertools 
from sktime.classification.kernel_based import RocketClassifier
import numpy as np
```

```python
machines = ["M01","M02","M03"]
process_names = ["OP00","OP01","OP02","OP03","OP04","OP05","OP06","OP07","OP08","OP09","OP10","OP11","OP12","OP13","OP14"]
labels = ["good","bad"]
path_to_dataset = Path("./data/").absolute()
```

```python
X_data = []
y_data = []

for process_name, machine, label in itertools.product(process_names, machines, labels):
    data_path = os.path.join(path_to_dataset, machine, process_name, label)
    data_list, data_label = data_loader_utils.load_tool_research_data(data_path, label=label, add_additional_label = False)
    #concatenating
    X_data.extend(data_list)
    y_data.extend(data_label)
```

```python
file = Path("./data/M01/OP00/good/M01_Aug_2019_OP00_000.h5").absolute()
sample = data_loader_utils.datafile_read(file)
```

```python
y_new = map(lambda x: 0 if x == "bad" else 1, y_data)
```

```python
y_map = list(y_new)
y = np.array(y_map)
```

```python
arrX = np.empty((0, 3,4096), dtype = np.float16)
arry = np.empty((0,))
for example in zip(X_data[:1000], y[:1000]):
    window_size = 4096
    modulo = X_data[0].shape[0] % 4096
    floor = X_data[0].shape[0] // 4096
    X_new = np.reshape(X_data[0][:-modulo,:],(floor, 3, 4096))
    arrX = np.concatenate((arrX, X_new), dtype = np.float16)
    if example[1] == 0:
        arry = np.concatenate((arry, np.zeros(floor)))
    else:
        arry = np.concatenate((arry, np.ones(floor)))
```

```python
subX = arrX[:10000,:,:]
suby = arry[:10000]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(subX, suby, test_size = 0.2)
```
